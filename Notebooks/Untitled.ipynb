```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Data/ufc_fighters.csv')
```

```python
df2 = pd.read_csv('../Data/UFC_Fighters_Database.csv')
```

```python
df['reach'] = df.reach.map(lambda x: str(x).split('"')[0])
```

```python
lists = ['Str_Acc', 'Str_Def', 'TD_Acc', 'TD_Def']
```

```python
for x in lists:
    df[x] = df[x].map(lambda m: int(str(m).split('%')[0])/100)
```

```python
def app(x):
    if x == "nan":
        return np.nan
    else:
        return x
```

```python
df.reach = df.reach.apply(app)
```

```python
for indx, row in df.iterrows():
    if row.reach is np.nan:
        j = df2[df2.NAME == df.name[row.name]]
        if len(j.REACH)>0:
            m = np.array(j.REACH)
            df.reach[row.name] = m[0]
        else:
            df.reach[row.name] = 0
            pass
```

```python
df.reach = df.reach.apply(lambda x: int(x))
```

```python
df.to_csv('../Data/fighters.csv', index=False)
```
