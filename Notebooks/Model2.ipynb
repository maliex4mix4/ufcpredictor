```python
df_copy = df.dropna()
```

```python
df_copy['target'] = df_copy['winner'].astype('str')+df_copy['method']+df_copy['end_round'].astype('str')
```

```python
df_copy.target.unique()
```

```python
unique_values = ['1KO/TKO4', '0KO/TKO3', '1KO/TKO1', '1SUB2', '0SUB1', '0DEC3',
       '1KO/TKO3', '1KO/TKO2', '1DEC3', '0SUB2', '0SUB3', '1SUB1',
       '0KO/TKO1', '0KO/TKO2', '0DEC5', '1DEC5', '1SUB3', '0SUB5', '0DQ1',
       '1DQ3', '0KO/TKO4', '1DQ1', '1SUB4', '1KO/TKO5', '0SUB4',
       '0KO/TKO5', '1DQ2', '1SUB5']
```

```python
def get_one_hot(c):
    return 1+unique_values.index(c)

df_copy['target'] = df_copy['target'].map(lambda x: get_one_hot(x))
```

```python
final_df = pd.get_dummies(df_copy.drop(['fighter1', 'fighter2', 'winner', 'method', 'end_round'], axis=1))
```

```python
final_df.target
```

```python
# Importng 

from sklearn.metrics import f1_score, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from catboost import CatBoostClassifier
```

```python
X = final_df.drop('target', axis=1)
Y = final_df['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.25, random_state=45)
```

```python
rfc = RandomForestClassifier(n_estimators=200, random_state=12)

rfc.fit(x_train, y_train)
rfc_pred = rfc.predict(x_test)

print(f"F1 Score for Random Forest is: {f1_score(y_test, rfc_pred, average='micro')} and accuracy Score is: {accuracy_score(y_test, rfc_pred)}")
```

```python
xgb = XGBClassifier(n_estimators=200, random_state=12)

xgb.fit(x_train, y_train)
xgb_pred = xgb.predict(x_test)

print(f'F1 Score for XGB is: {f1_score(y_test, xgb_pred, average="micro")} and accuracy Score is: {accuracy_score(y_test, xgb_pred)}')
```

```python
lr = LogisticRegression()

lr.fit(x_train, y_train)
lr_pred = lr.predict(x_test)

print(f'F1 Score for LR is: {f1_score(y_test, lr_pred, average="micro")} and accuracy Score is: {accuracy_score(y_test, lr_pred)}')
```

```python
cat = CatBoostClassifier(n_estimators=200, random_state=12)

cat.fit(x_train, y_train)
cat_pred = cat.predict(x_test)

print(f'F1 Score for CAT boost is: {f1_score(y_test, cat_pred, average="micro")} and accuracy Score is: {accuracy_score(y_test, cat_pred)}')
```
